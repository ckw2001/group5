```python
import json 
from Binomial_American import binomial_tree_american_option
import matplotlib.pylab as plt
```

### How Strike Price Affect Call and Put Option Price

```python
S = 50
t = 0
T = 0.5
sigma = 0.2
r = 0.01
N = 200
price_call = []
price_put = []
K_list = []
for K in range(10,100,10):
    
    C = float(json.loads(binomial_tree_american_option(S, sigma, r, T, K, N, option_type='call'))['price'])
    P = float(json.loads(binomial_tree_american_option(S, sigma, r, T, K, N, option_type='put'))['price'])
    price_call.append(C)
    price_put.append(P)
    K_list.append(K)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
# the first subplot
ax1.plot(K_list,price_call,'o-',label='call')
ax1.set_title('American call option')
ax1.set_xlabel('Strike price')
ax1.set_ylabel('Call option price')
ax1.legend()
# the second subplot
ax2.plot(K_list,price_put,'o-',label = 'put')
ax2.set_title('American put option')
ax2.set_xlabel('Strike price')
ax2.set_ylabel('Put option price')
ax2.legend()
# adjust layout and show
plt.tight_layout()
plt.show()
```

### How Risk Free Rate Affects Call and Put Option Price

```python
S = 50
K =50
t = 0
T = 0.5
sigma = 0.2
N = 200
price_call = []
price_put = []
variable_list = []
for r in range(1,40,1):
    r= r*0.01
    C = float(json.loads(binomial_tree_american_option(S, sigma, r, T, K, N, option_type='put'))['price'])
    P = float(json.loads(binomial_tree_american_option(S, sigma, r, T, K, N, option_type='call'))['price'])
    price_call.append(C)
    price_put.append(P)
    variable_list.append(r)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
# the first subplot
ax1.plot(variable_list,price_call,'o-',label='call')
ax1.set_title('American call option')
ax1.set_xlabel('Interest rate')
ax1.set_ylabel('Call option price')
ax1.legend()
# the second subplot
ax2.plot(variable_list,price_put,'o-',label = 'put')
ax2.set_title('American put option')
ax2.set_xlabel('Interest rate')
ax2.set_ylabel('Put option price')
ax2.legend()
# adjust layout and show
plt.tight_layout()
plt.show()
```
